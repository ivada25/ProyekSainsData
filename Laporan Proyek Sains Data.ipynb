# 1. pendahuluan

## Latar Belakang

Kesuburan tanah merupakan salah satu faktor penting dalam pertanian untuk menentukan tingkat keberhasilan panen. Untuk bisa menentukan tingkat kesuburan sebuah tanah bisa melalui beberapa indikator yaitu tingkat Ph tanah, kandungan unsur hara tanah, tekstur tanah, warna tanah, dan masih banyak indikator lainnya penentu kesuburan tanah. Tanah yang subur bisa menyediakan unsur hara yang dibutuhkan oleh tanaman dalam kadar yang cukup. Namun, ada beberapa tanah yang memiliki tingkat kesuburan yang kurang baik sehingga membutuhkan bantuan melalui pupuk untuk bisa membantu proses pertumbuhan tanaman agar menghasilkan panen dengan kualitas yang baik dan bagus. Meskipun dengan pemberian pupuk bisa membantu tanah untuk bisa membantu proses pertumbuhan tanaman, pemberian pupuk tersebut juga harus sesuai dengan kadar yang seharusnya, karena jika kadar pupuk yang diberikan kurang atau lebih dari kadar seharusnya, hal tersebut juga bisa menghambat pertumbuhan tanaman.

### Identifikasi Outlier

```python
import pandas as pd
from sklearn.neighbors import LocalOutlierFactor

# Baca file CSV soil_fertility.csv
file_path = 'soil fertility.csv'  # Ganti dengan path file kamu
soil = pd.read_csv(file_path)

# Simpan dataframe ke df_wine (sesuai dengan kode asli)
soil_fertility = soil.copy()

# Simpan kolom 'quality'
wine_quality = soil_fertility[['Output']]

# Hapus kolom 'quality' karena tidak digunakan dalam model LOF
soil_fertility = soil_fertility.drop(['Output'], axis=1)

# Inisialisasi dan fit model LOF
lof = LocalOutlierFactor(n_neighbors=4, contamination=0.1)  # Set parameter LOF
outlier_labels = lof.fit_predict(soil_fertility)

# Tambahkan label outlier dan kolom 'quality' kembali ke dataframe
soil_fertility['Outlier'] = outlier_labels
soil_fertility['Output'] = wine_quality

# Data yang sudah dibersihkan, tanpa outlier (optional)
df_cleaned = soil_fertility[soil_fertility['Outlier'] != -1]

# Tampilkan dataframe dengan label outlier dan kolom quality
soil_fertility
```
