# Laporan Mahamami Bisnis 

# Prediksi Tingkat Kesuburan Tanah Untuk Pemberian Dosis Pupuk

# 1. pendahuluan

## Latar Belakang

### visi dan misi

Visi 
Menjadi pusat penelitian kualitas tanah untuk mendukung pengembangan dan meningkatkan hasil panen dari sektor pertanian.
Misi
- Meningkatkan pemahaman tentang kualitas tanah untuk keberhasilan pertanian 
- Membangun kesadaran masyarakat terkait pengaruh kualitas tanah terhadap proses penanaman
- Berkolaborasi dengan pemerintah dan lembaga sektor pertanian untuk menyediakan sarana dan prasarana agar bisa mendukung kebijakan yang sudah direncanakan.

### Strategi Bisnis

Strategi bisnis yang dilakukan oleh peneliti yaitu membangun prediksi terkait tingkat kesuburan tanah yang akan digunakan untuk media tanam agar mengetahui dosis pemberian pupuk yang sesuai dengan kondisi tanah yang tersedia agar meningkatkan hasil panen yang baik berdasarkan hasil prediksi tersebut. Setelah melakukan pemrediksian tersebut nantinya hasilnya akan di koordinasikan dengan sektor industri yang terkait dan juga masyarakat umum yang membutuhkan informasi tersebut untuk kepentingan usahanya, agar tingkat hasil panen yang memiliki kualitas terbaik di Indonesia semakin bertambah dengan adanya informasi tersebut. 

### Target Bisnis

Karena penelitian ini bergerak di bidang pertanian dimana untuk mengecek dan melakukan pemrediksian kualitas tanah untuk menentukan dosis pemberian pupuk yang sesuai dengan kondisi tanah tersebut maka peneliti menargetkan bisnis ini untuk kalangan industri pertanian dan juga masyarakat umum yang memiliki usaha dalam bidang pertanian.

### Tujuan Bisnis

Kualitas tanah merupakan faktor penting dalam industri pertanian karena setiap tanah memiliki Tingkat kesuburan yang berbeda. Berdasarkan Analisa kesuburan tanah tersebut nantinya dilakukan prediksi untuk pemberian jumlah pupuk yang sesuai dengan kondisi tanah tersebut. Hal tersebut merupakan tujuan dari adanya penelitian ini, yaitu untuk mengetahui dosis pemberian pupuk pada tanah berdasarkan Tingkat kesuburan tanah untuk menghindari adanya kelebihan ataupun kekurangan dosis pemberian pupuk pada suatu tanah agar bisa meminimalisir terjadinya gagal panen pada suatu tanaman.
Maka diperlukan adanya analisa terhadap kesuburan tanah tersebut untuk mengetahui jumlah pemberian pupuk yang tepat terhadap tanah tersebut agar bisa menghasilkan panen yang bagus dan baik pada tanaman tersebut.

### Rumusan Masalah

Masalah yang bisa muncul dalam penelitian ini yaitu terkait dengan sumber daya manusia yang ada, karena tidak sedikit masyakarat umum yang meremehkan terkait adanya penelitian seperti ini dikarenakan pemikiran mereka yang cenderung tidak peduli terkait perubahan yang bisa terjadi dengan adanya penelitian yang dilakukan oleh peneliti dengan tujuan untuk meningkatkan nilai panen.
